```python
from functools import reduce
from itertools import chain

k1 = np.array([1, 2, 3])
k2 = np.array([4, 5, 6])
k3 = np.array([4, 5, 6, 7, 'apple', 'dinosaur', '⭐⭐⭐⭐⭐'])

# np.concatenate((k1, k2, k3))
# np.array([k1, k2]).flatten()

# np.hstack((k1, k2, k3))
# np.vstack((k1, k2))

# k3.dtype

tup = (k1.tolist(), k2.tolist(), k3.tolist())
list(chain.from_iterable(tup))
reduce(lambda acc, k: acc + k, tup, [])
sum(tup, [])
```
